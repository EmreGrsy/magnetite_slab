```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
font = {"size"   : 20}
plt.rc("font", **font)
plt.rcParams['ytick.right']

dftet = pd.read_csv("nfe_tet.data", skiprows = 0, 
                 header = None, 
                 delimiter=" ")

dfoct = pd.read_csv("nfe_oct.data", skiprows = 0, 
                 header = None, 
                 delimiter=" ")
```

```python
dftet[1]/(dftet[2]+dftet[1])
```

```python
fig, axis = plt.subplots(1, 2, figsize=(5, 6), facecolor="w", 
                         sharex=True, sharey=True)

fig.subplots_adjust(wspace=0.1)

minor_ticks = np.arange(1, 28, 1)
major_ticks = np.arange(1, 28, 5)

axis[0].set_yticks(major_ticks)
axis[0].set_yticks(minor_ticks, minor=True)
axis[0].grid(which='minor', alpha=0.5, linestyle='dotted')
axis[0].grid(which='major', alpha=0.5, linestyle='dotted')

axis[1].set_yticks(major_ticks)
axis[1].set_yticks(minor_ticks, minor=True)
axis[1].grid(which='minor', alpha=0.5, linestyle='dotted')
axis[1].grid(which='major', alpha=0.5, linestyle='dotted')

axis[1].set_xticks([0, 0.5, 1.0])
axis[0].set_xlim([-0.15, 1.15])

axis[0].invert_yaxis()

axis[0].set_ylabel(r"Fe$_\mathrm{tet}$ layer", ha="center", fontsize=24, labelpad=20)
axis[1].set_ylabel(r"Fe$_\mathrm{oct}$ layer", ha="center", fontsize=24, labelpad=20)

axis[0].yaxis.set_label_position("left")
axis[1].yaxis.set_label_position("right")

#axis[0].yaxis.grid(color='gray', linestyle='dotted', alpha=0.5)
#axis[0].xaxis.grid(color='gray', linestyle='dotted', alpha=0.5)

axis[0].plot(dftet[1]/(dftet[2]+dftet[1]), dftet[0], linewidth=4, color="blue")

axis[1].plot(dfoct[1]/(dfoct[2]+dfoct[1]), dfoct[0], linewidth=4, color="blue")

ax0 = fig.add_subplot(111, frame_on=False)
ax0.set_xticks([])
ax0.set_yticks([])
ax0.set_xlabel(r"$n_{\mathrm{{Fe}_{oct}^{3+}}}/\mathrm{n_{{Fe}_{oct}}}$", labelpad=35, fontsize=24)
ax0.set_title(r"(111)-tet, 41 layer", pad=20)
#ax0.set_title("21L + 24L", fontsize=30, pad=55)
plt.show()
```

```python
fig.savefig('111-tet-nfe3.pdf',format='pdf', bbox_inches = "tight")
fig.savefig('111-tet-nfe3.png', dpi=300.0,format='png', bbox_inches = "tight")
```
